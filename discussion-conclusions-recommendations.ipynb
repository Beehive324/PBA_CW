```python
import matplotlib.pyplot as plt
import numpy as np

models = ['KNN', 'Naive Bayes', 'Decision Tree', 'ANN', 'SVM', 'Logistic Regression', 
          'Random Forest']
metrics = ['Accuracy', 'ROC AUC']

#Accuracies and ROC AUC for each model 
scores = np.array([
    [0.88, 0.50], #KNN
    [0.88, 0.55], #Naive Bayes
    [0.89, 0.55], #Decision Tree  
    [0.88, 0.50], #ANN
    [0.88, 0.55], #SVM
    [0.88, 0.50], #LR
    [0.90, 0.94] #RF
])

#Bar Chaart
bar_width = 0.3
x = np.arange(len(models))

plt.figure(figsize=(10, 6))
for i, metric in enumerate(metrics):
    plt.bar(x + i * bar_width, scores[:, i], width=bar_width, label=metric)

plt.xlabel('Models', fontsize=10)
plt.ylabel('Scores', fontsize=10)
plt.title('Model Performance Metrics', fontsize=10)
plt.xticks(x + bar_width * (len(metrics) - 1) / 2, models) 
plt.ylim(0, 1) 
plt.legend(title='Metrics')
plt.tight_layout()
plt.show()
```
